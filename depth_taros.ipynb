```python

import os

# Specify the path to the folder containing the images
folder_path = '/Users/liiuhaoran/Downloads/Scaffold-GS-main/gp_vis/lego/train'

# Get all the files in the folder
file_list = os.listdir(folder_path)

# Filter to include only files that have a .png extension (ignore other file types)
image_files = [file for file in file_list if file.endswith('.png')]

# Sort the files based on their current names
image_files.sort()

# Rename files starting from 000000.png
for idx, old_filename in enumerate(image_files):
    # Create the new filename with zero-padding
    new_filename = f"{idx:06}.png"

    # Construct full file paths
    old_file_path = os.path.join(folder_path, old_filename)
    new_file_path = os.path.join(folder_path, new_filename)

    # Rename the file
    os.rename(old_file_path, new_file_path)

print(f"Renamed {len(image_files)} files in {folder_path}")
```
